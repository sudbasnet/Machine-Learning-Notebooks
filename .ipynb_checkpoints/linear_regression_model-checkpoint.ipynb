### create polynomial features of degree n

```python
def polyFeatures1D(x, degree):
    polyFeat1D, deg = np.array([1.0]), 1
    startpoint = np.zeros(len(x))
    while deg <= degree:
        polyFeat1D_ = np.array([])
        for i in range(len(x)):
            curr = x[i]
            y = polyFeat1D[int(startpoint[i]):]
            startpoint[i] = (len(polyFeat1D) + len(polyFeat1D_))
            polyFeat1D_ = np.append(polyFeat1D_, curr * y)
        polyFeat1D = np.append(polyFeat1D, polyFeat1D_)
        deg += 1
    return polyFeat1D

def polynomialFeatures(X, degree=1):
    polyFeat = np.array([])
    for i in range(len(X)):
        if len(polyFeat) == 0:
            polyFeat = [polyFeatures1D(X[i], degree)]
        else:        
            polyFeat = np.append(polyFeat, [polyFeatures1D(X[i], degree)], axis = 0)
    return polyFeat
```

```python
# error function
def mse(Y_true, Y_pred):
    if(len(Y_pred) == len(Y_true)):
        loss = Y_pred - Y_true
        cost = np.sum(loss ** 2) / (2 * len(Y_pred))
    return cost
```

```python
def learning_curve(model, X, Y, cv, train_size=1, 
                   learning_rate=0.01, epochs=1000, tol=None, 
                   regularizer=None, lambd=0.0, **kwargs):
    train_scores, val_scores = [], []
    sample_size = train_size
    model_p = {"learning_rate": learning_rate, "lambd": lambd, 
     "regularizer": regularizer, "epochs": epochs, "tol": tol}
    
    while sample_size <= len(X):
        x, y = X[:sample_size], Y[:sample_size]
        data = kFold(cv, x, y, model, model_p, mse)
        train_scores.append( np.sqrt(data["average_error"]["error_train"]))
        val_scores.append( np.sqrt(data["average_error"]["error_validation"])) 
        sample_size +=  train_size
        
    return val_scores, train_scores
```

```python
def plot_polynomial_model_complexity(model, X, Y, cv, 
                                     maxPolynomialDegree, 
                                     learning_rate=0.01, epochs=1000, 
                                     tol=None, regularizer=None, 
                                     lambd=0.0, **kwargs):
    mse_train, mse_test = [], []
    for degree in range(1, maxPolynomialDegree + 1):
        X_ = polynomialFeatures(X, degree)
        model_p = {"learning_rate": learning_rate, "lambd": lambd,
                   "regularizer": regularizer, "epochs": epochs, 
                   "tol": tol}
        data = kFold(cv, X_, Y, model, model_p, mse)
        mse_train.append(data["average_error"]["error_train"])
        mse_test.append(data["average_error"]["error_validation"])
    print(mse_train)
    print(mse_test)
    plt.figure(figsize=(10, 6))   
    plt.plot(range(1, maxPolynomialDegree + 1), mse_test, "ro-", alpha=1.0, linewidth=1.0, label="Test RMSE")
    plt.plot(range(1, maxPolynomialDegree + 1), mse_train, "bo-", alpha=1.0, linewidth=1.0, label="Train RMSE")    
    plt.legend(loc="best", fontsize=14) 
    plt.xlabel("Degree")
    plt.ylabel("RMSE")
    plt.title("RMSE for Varying Degree")
    plt.show()
        
```

```python
class Linear_Regression:
    def __init__(self):
        pass
    
    def fit(self, X, Y, learning_rate=0.01, epochs=1000, tol=None, regularize=None, lambd=0.0, **kwargs):
        # add bias
        X = np.insert(X, 0, 1, axis=1)
        # initial theta values = 0
        theta_hat = np.zeros(X.shape[1])
        # number of samples
        size = X.shape[0] * 1.0 
        # initial err values
        err = mse(Y, np.dot(X, theta_hat))
        
        for epoch in range(epochs):
            hypothesis = np.dot(X, theta_hat)
            
            # if tol is not None, and if it is greater than previous error  - tol then break
            if tol and mse(Y, hypothesis) > (err - tol):
                break
            else:
                err = mse(Y, hypothesis)
            
            theta_grad = np.dot(X.transpose(),(hypothesis - Y))/(size * 1.0)
            if regularizer == "None":    
                # gradient descent update role
                theta_hat = theta_hat - (learning_rate * theta_grad)
            else:
                reg_theta = None
                # exclude bias term
                temp_theta = np.delete(theta_hat, 0)
                if regularizer == "l1":
                    # add bias back
                    reg_theta = np.concatenate(([0], np.sign(temp_theta)))
                elif regularizer == "l2":
                    # add bias back
                    reg_theta = np.concatenate(([0], temp_theta))
                # regularization
                theta_hat = theta_hat - (learning_rate * theta_grad) - ((learning_rate * lambd * reg_theta) / (size * 1.0))
        # update theta  
        self.theta = theta_hat
        
    def predict(self, X):
        # add bias
        X = np.insert(X, 0, 1, axis=1)
        hypothesis = np.dot(X, self.theta)
        return hypothesis
```

```python
# partition function generate X_train, X_test, y_train, y_test
def partition(feature_arr, target_arr, t=0.2):
    nrows = int(np.floor((1-t) * feature_arr.shape[0]))
    X_train, X_test  = np.array(feature_arr[:nrows]), np.array(feature_arr[nrows:])
    y_train, y_test = np.array(target_arr[:nrows]), np.array(target_arr[nrows:])
    return X_train, X_test, y_train, y_test
```

### Model Evalulation

```python
def kFold(folds, data, labels, model, model_args, error_function):
    error_train = []
    error_validation = []
    nrows = data.shape[0]
    # we divide the dataset in equal parts. If there are extra rows, we ignore them
    f_size = int(np.floor(nrows/folds))
    data, labels = data[0:(f_size * folds)], labels[0:(f_size * folds)]
    data, labels = np.split(data, folds), np.split(labels, folds)

    expected_labels = np.ndarray([])
    predicted_labels = np.ndarray([])
    
    for fold in range(folds):
        test_created = False
        for i in range(folds):
            if fold == i:
                X_test = data[i]
                y_test = labels[i]
                test_created = True
            else:
                if i==0 or (i == 1 and test_created):
                    X_train = data[i]
                    y_train = labels[i]
                else:
                    X_train = np.append(X_train, data[i], axis=0)
                    y_train = np.append(y_train, labels[i], axis=0)
        # test and train sets have been created
        
        linear_regression = model()
        linear_regression.fit(X_train, y_train, learning_rate=model_args['learning_rate'], lambd=model_args['lambd'], regularizer=model_args['regularizer'])
        
        y_pred_train = linear_regression.predict(X_train)
        y_pred_validation = linear_regression.predict(X_test)
            
        expected_labels = np.append(expected_labels, y_test)
        predicted_labels = np.append(predicted_labels, y_pred_validation)
        error_train.append(error_function(y_pred_train, y_train))
        error_validation.append(error_function(y_pred_validation, y_test))
        error = {'error_train': np.mean(error_train), 'error_validation': np.mean(error_validation)}
#         print("error_train: ", np.mean(error_train), ", error_validation: " , error_validation)
    return {'expected_labels': expected_labels, 'predicted_labels': predicted_labels, 'average_error': error}
```

```python
def scale_features(features):
    mean_train = np.mean(features, axis=0)
    std_train = np.std(features, axis=0)
    return (features - mean_train)/(std_train * 1.0)

features = df.loc[:, df.columns != 'quality'].values
features = scale_features(features)

target = df['quality'].values

eval_results = []
column_names = ['lambd', 'learning_rate', 'regularizer', 'mse_train', 'mse_validation', 'mse_average']
model_eval = pd.DataFrame(columns=column_names)

for lambd in [1.0, 0.0, 0.1, 0.01, 0.001, 0.0001]:
    for l_rate in [1.0, 0.1, 0.01, 0.001, 0.0001]:
        for regularizer in ['l1', 'l2']:
#             plot_polynomial_model_complexity(Linear_Regression, features, target, 5, 4, learning_rate=l_rate, epochs=1000, tol=None, regularizer=regularizer, lambd=lambd)        
            kFold_results = kFold(8, features, target, Linear_Regression, {'lambd': lambd, 'learning_rate': l_rate, 'regularizer': regularizer}, mse)
            avg_error = (kFold_results['average_error']['error_train'] + kFold_results['average_error']['error_validation'])/2.0
            tempDF = pd.DataFrame([[lambd, l_rate, regularizer, kFold_results['average_error']['error_train'], kFold_results['average_error']['error_validation'], avg_error]], columns=column_names)
            model_eval = model_eval.append(tempDF, ignore_index = True)

model_eval.sort_values('mse_average')
print(model_eval)
```

```python
# Evaluate on test data
features = df.loc[:, df.columns != 'quality'].values
features = scale_features(features)
target = df['quality'].values
X_train, X_test, y_train, y_test = partition(features, target, t=0.2)

lin_reg = Linear_Regression()
lin_reg.fit(X_train, y_train, learning_rate=0.01, lambd=0.0001, regularizer='l1')
y_test_pred = lin_reg.predict(X_test)
print("Mean squared error: %.2f" % mse(y_test, y_test_pred))
```

```python
# Plot rmse for learning_curve
train_size = 10
train_scores, val_scores = learning_curve(Linear_Regression, features, target, 5, train_size=train_size, learning_rate=0.01, epochs=1000, tol=None, regularizer=None, lambd=0.0)

size = features.shape[0]
train_sizes_xaxis = np.arange(train_size, size, train_size)

plt.figure(figsize=(10, 6))

plt.plot(train_sizes_xaxis, train_scores, "r-+", linewidth=3, label="Training Score")
plt.plot(train_sizes_xaxis, val_scores, "b-", linewidth=2, label="Cross-validation Score")
plt.legend(loc="best", fontsize=14)   
plt.xlabel("Training set size", fontsize=14) 
plt.ylabel("Negative MSE", fontsize=14) 
plt.title("Learning Curve")
plt.show()
```
